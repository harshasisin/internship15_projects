```python
#import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementClickInterceptedException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
```

# Data Collection Phase

```python
#creating the empty list
airline_name=[]
date_of_journey=[]
source=[]
destination=[]
route=[]
departure_time=[]
arrival_time=[]
duration=[]
total_stops=[]
price = []
```

```python
#Go to yatra.com website
driver=webdriver.Chrome(r"C:\Users\sisin\chromedriver_win32\chromedriver.exe") 
time.sleep(4)
url = "https://www.yatra.com/"
driver.get(url)
time.sleep(4)
```

```python
urls = []
```

```python
#collect urls of cities
lis = driver.find_elements_by_xpath("//ul[@class='destlist'][1]/li")
for i in lis:    
    i.click()
    time.sleep(3)
    slides = driver.find_elements_by_xpath("//a[@class='VueCarousel-slide items']")
    for i in slides:
        urls.append(i.get_attribute('href'))
    urls    
```

```python
url1 = []
url2 = []
```

```python
for i in urls:
    if i.find('flight-schedule') == -1:
        url1.append(i)
    else :
        url2.append(i)
```

```python
url2
```

```python
#collect  data from each url
for i in url2:
    driver.get(i)
    time.sleep(5)
    try:
        for i in driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']"):
            airline_name.append(i.text)
            date_of_journey.append(driver.find_element_by_xpath("//div[@class='day-li text-center cursor-pointer pr active font-primary-color']/p[1]").text)
    except NoSuchElementException:
        airline_name.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p"):
            source.append(i.text)
    except NoSuchElementException:
        source.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p"):
            route.append(i.text)
    except NoSuchElementException:
        route.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]"):
            destination.append(i.text)
    except NoSuchElementException:
        destination.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='i-b pr']"):
            departure_time.append(i.text)
    except NoSuchElementException:
        departure_time.append("--")
    try:
        for i in driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']"):
            arrival_time.append(i.text)
    except NoSuchElementException:
        arrival_time.append("--")
    try:
        for i in driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']"):
            duration.append(i.text)
    except NoSuchElementException:
        duration.append("--")
    try:
        for i in driver.find_elements_by_xpath("//span[@class=' font-lightgrey fs-11 tipsy i-b']/span"):
            total_stops.append(i.text)
    except NoSuchElementException:
        total_stops.append("--")
    try:
        for i in driver.find_elements_by_xpath("//p[@class='i-b']"):
            price.append(i.text)
    except NoSuchElementException:
        price.append("--")
```

```python
#collect  data from each url
for i in url1:
    driver.get(i)
    time.sleep(5)
    try:
        for i in driver.find_elements_by_xpath("//span[@class='text i-b']"):
            airline_name.append(i.text)
            date_of_journey.append(driver.find_element_by_xpath("//span[@class='depart-date-span date-span']").text)
    except NoSuchElementException:
        airline_name.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='fs-16 bold mb-2 time']/p[2]"):
            source.append(i.text)
    except NoSuchElementException:
        source.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p"):
            route.append(i.text)
    except NoSuchElementException:
        route.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='i-b col-5 pdd-0 text-left atime']/p[2]"):
            destination.append(i.text)
    except NoSuchElementException:
        destination.append("--")
    try:
        for i in driver.find_elements_by_xpath("//p[@class='fs-15 text-uppercase font-lightestgrey normal mr-2 ccode hide-mobile']"):
            departure_time.append(i.text)
    except NoSuchElementException:
        departure_time.append("--")
    try:
        for i in driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']"):
            arrival_time.append(i.text)
    except NoSuchElementException:
        arrival_time.append("--")
    try:
        for i in driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']"):
            duration.append(i.text)
    except NoSuchElementException:
        duration.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-11 tipsy i-b']/span"):
            total_stops.append(i.text)
    except NoSuchElementException:
        total_stops.append("--")
    try:
        for i in driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']"):
            price.append(i.text)
    except NoSuchElementException:
        price.append("--")
```

```python
#collect  data from each url
for i in url2:
    driver.get(i)
    time.sleep(3)
    for i in driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-11 tipsy i-b']"):
        try:
            total_stops.append(i.find_element_by_xpath(".//span[@class='dotted-borderbtm']").text)
        except NoSuchElementException:
            total_stops.append('--')
```
